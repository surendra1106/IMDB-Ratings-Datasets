### Univariate Analysis - Numerical Features

```python
num_cols = ['Year','Runtime (Minutes)','Votes','Revenue (Millions)','Metascore']
for col in num_cols:
    if col in df.columns:
        plt.figure(figsize=(8,4))
        sns.histplot(df[col].dropna())
        plt.title(f"Distribution of {col}")
        plt.show()
```

### Univariate Analysis - Categorical Features

```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_cols = ['Genre', 'Director']

for col in cat_cols:
    if col in df.columns:
        top_vals = df[col].value_counts().head(10)
        plt.figure(figsize=(10,5))
        sns.barplot(x=top_vals.values, y=top_vals.index, color='skyblue')
        plt.title(f"Top 10 {col} values")
        plt.xlabel('Count')
        plt.ylabel(col)
        plt.show()
```

### Features vs Target (Bivariate Analysis)

```python
# Numerical vs Target
num_cols = ['Votes', 'Revenue (Millions)', 'Metascore', 'Runtime (Minutes)']

for col in num_cols:
    if col in df.columns:
        sns.scatterplot(data=df, x=col, y='Rating', alpha=0.6)
        plt.title(f"{col} vs Rating")
        plt.show()
```

```python
# Categorical vs Target
if 'Genre' in df.columns:
    top_genres = df['Genre'].value_counts().head(5).index
    sns.boxplot(
        data=df[df['Genre'].isin(top_genres)],
        x='Genre',
        y='Rating'
    )
    plt.xticks(rotation=45)
    plt.title("Rating by Genre (Top 5)")
    plt.show()
```

### Categorical vs Categorical (Example)

```python
df['Genre'] = df['Genre'].str.split(',')
df['Genre'] = df['Genre'].apply(lambda x: [g.strip() for g in x])

# 2 Explode Genre into multiple rows
df_exploded = df.explode('Genre')

# 3 Filter top categories
top_genres = df_exploded['Genre'].value_counts().head(5).index
top_directors = df['Director'].value_counts().head(5).index
df_filtered = df_exploded[df_exploded['Genre'].isin(top_genres) & df_exploded['Director'].isin(top_directors)]

# 4 Create crosstab
cross_tab = pd.crosstab(df_filtered['Genre'], df_filtered['Director'])

# 5 Plot stacked bar chart
cross_tab.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')
plt.title('Genre vs Director (Stacked, Top 5 each)')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Director')
plt.show()
```

```python
# Numerical columns
num_cols = ['Votes', 'Revenue (Millions)', 'Metascore', 'Runtime (Minutes)']

# Compute correlation with target
corr_target = df[num_cols + ['Rating']].corr()[['Rating']].drop('Rating')  # keep only features vs target

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_target, annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation of Numerical Features with Rating')
plt.show()
```
